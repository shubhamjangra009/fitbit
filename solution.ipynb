```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

## Daily Activity Data

```python
DailyCalories = pd.read_csv('OriginalData/dailyCalories_merged.csv', parse_dates=['ActivityDay'])
DailyIntensities = pd.read_csv('OriginalData/dailyIntensities_merged.csv', parse_dates=['ActivityDay'])
DailySteps = pd.read_csv('OriginalData/dailySteps_merged.csv', parse_dates=['ActivityDay'])
```

## Hourly Activity Data

```python
format='%m/%d/%Y %I:%M:%S %p'
HourlySteps = pd.read_csv('OriginalData/hourlySteps_merged.csv', parse_dates=['ActivityHour'], date_format=format)
HourlyCalories = pd.read_csv('OriginalData/hourlyCalories_merged.csv', parse_dates=['ActivityHour'], date_format=format)
HourlyIntensities = pd.read_csv('OriginalData/hourlyIntensities_merged.csv', parse_dates=['ActivityHour'], date_format=format)
```

## Minutes Activity Data

```python
MinuteSteps = pd.read_csv('OriginalData/minuteStepsNarrow_merged.csv', parse_dates=['ActivityMinute'], date_format=format)
MinuteCalories = pd.read_csv('OriginalData/minuteCaloriesNarrow_merged.csv', parse_dates=['ActivityMinute'], date_format=format)
MinuteIntensities = pd.read_csv('OriginalData/minuteIntensitiesNarrow_merged.csv', parse_dates=['ActivityMinute'], date_format=format)
MinuteMETs = pd.read_csv('OriginalData/minuteMETsNarrow_merged.csv', parse_dates=['ActivityMinute'], date_format=format)
MinuteSleep = pd.read_csv('OriginalData/minuteSleep_merged.csv', parse_dates=['date'], date_format=format)
```

## Heart Rate, Sleep Data, Weight Log Data

```python
Heartrate = pd.read_csv('OriginalData/heartrate_seconds_merged.csv', parse_dates=['Time'], date_format=format)
Sleep = pd.read_csv('OriginalData/sleepDay_merged.csv', parse_dates=['SleepDay'], date_format=format)
WeightLog = pd.read_csv('OriginalData/weightLogInfo_merged.csv', parse_dates=['Date'], date_format=format)
```

## __FINAL DATASETS__
---

## 1. Daily dataset 

```python
Daily = pd.merge(DailySteps, DailyCalories, how='outer', on=['Id', 'ActivityDay'])
Daily = pd.merge(Daily, DailyIntensities, how='outer', on=['Id', 'ActivityDay'])
# Daily.ActivityDay = Daily.ActivityDay.dt.strftime('%d/%m/%Y')
Daily
```

## 2. Hourly dataset

```python
Hourly = pd.merge(HourlySteps, HourlyCalories, how='outer', on=['Id', 'ActivityHour'])
Hourly = pd.merge(Hourly, HourlyIntensities, how='outer', on=['Id', 'ActivityHour'])
# Hourly.ActivityHour = Hourly.ActivityHour.dt.strftime('%d/%m/%Y %H:%M')
Hourly
```

## 3. Minute dataset

```python
Minute = pd.merge(MinuteSteps, MinuteCalories, how='outer', on=['Id', 'ActivityMinute'])
Minute = pd.merge(Minute, MinuteIntensities, how='outer', on=['Id', 'ActivityMinute'])
Minute = pd.merge(Minute, MinuteMETs, how='outer', on=['Id', 'ActivityMinute'])
Minute.rename(columns={'Steps': 'TotalSteps'}, inplace=True)
Minute
```

Storing all datasets in a list

```python
all_datasets = [Daily, Hourly, Minute, Heartrate, Sleep, WeightLog]
```

Changing all Id columns of all datasets to string dtype

```python
for dataset in all_datasets:
    dataset.Id = dataset.Id.astype(str)
    dataset.rename(columns={'StepTotal': 'TotalSteps'}, inplace=True)
```

```python
Heartrate.rename(columns={'Time': 'DateTime'}, inplace=True)
WeightLog.rename(columns={'Date': 'DateTime'}, inplace=True)
```

Appending total distance, total active minutes, and activity day name cols at end

```python
Daily = Daily.assign(
    TotalActiveMinutes = Daily.LightlyActiveMinutes + Daily.FairlyActiveMinutes + Daily.VeryActiveMinutes,
    TotalDistance = Daily.LightActiveDistance + Daily.ModeratelyActiveDistance + Daily.VeryActiveDistance,
    ActivityDayName = Daily.ActivityDay.dt.day_name()
)
Daily
```

All individual participant stats

```python
DailyAggregatedData = round(pd.DataFrame(Daily.groupby('Id').agg(AverageStepsPerDay = ('TotalSteps', 'mean'),
                                MaxStepsOnADay = ('TotalSteps', 'max'),
                                NetTotalSteps = ('TotalSteps', 'sum'),
                                AverageDistancePerDay = ('TotalDistance', 'mean'),
                                MaxDistanceOnADay = ('TotalDistance', 'max'),
                                NetDistanceTravelled = ('TotalDistance', 'sum'),
                                AverageCalorieBurntPerDay = ('Calories', 'mean'),
                                MaxCalorieOnADay = ('Calories', 'max'),
                                NetCaloriesBurnt = ('Calories', 'sum')
                                )))
DailyAggregatedData
```

```python
print(f'Max Average Steps per Day: {DailyAggregatedData.AverageStepsPerDay.max()} by participant ID: {DailyAggregatedData.index[np.where(DailyAggregatedData.AverageStepsPerDay == DailyAggregatedData.AverageStepsPerDay.max())[0][0]]}')

print(f'Max Steps on a Day: {DailyAggregatedData.MaxStepsOnADay.max()} by participant ID: {DailyAggregatedData.index[np.where(DailyAggregatedData.MaxStepsOnADay == DailyAggregatedData.MaxStepsOnADay.max())[0][0]]}')

print(f'Max Net Total Steps per Day: {DailyAggregatedData.NetTotalSteps.max()} by participant ID: {DailyAggregatedData.index[np.where(DailyAggregatedData.NetTotalSteps == DailyAggregatedData.NetTotalSteps.max())[0][0]]}')
```

Boxplots

```python
fig, axes = plt.subplots(2, 1, figsize=(8,8))
fig.tight_layout()
fig.subplots_adjust(hspace=0.35)
ax1 = sns.boxplot(Daily, x='TotalSteps', ax=axes[0])
ax1.set_title('Average range of Total Steps by participants with some exceptions', loc='left');

ax2 = sns.boxplot(Daily, x='TotalDistance', ax=axes[1])
ax2.set_title('Average range of Total Distance travelled by participants with some exceptions', loc='left');
```

```python
corr_matrix = Daily.corr(numeric_only=True)
for i in range(corr_matrix.shape[0]):
    corr_matrix.iat[i, i] = np.nan
    
plt.figure(figsize=(10,8))
ax = sns.heatmap(round(corr_matrix, 1), annot=True, vmin=-1, vmax=1, cmap='coolwarm_r')
plt.title('CORRELATION MATRIX', fontsize=18, fontweight='bold', y=1.02);
plt.grid(False)
```
